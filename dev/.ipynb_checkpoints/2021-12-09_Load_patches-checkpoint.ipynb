```python
from aprovhots import get_labels_indices, fit_MLR
path = '/home/INT/grimaldi.a/Documents/projets/WP3/2021-12-06_simulator_data/'
#mean_isi = (isi_OFF+isi_ON)/2
patch_size = 5
mean_isi = 190*1e3
tau_cla = mean_isi*patch_size**2*1e-3 # tau has to be given in ms 
```

```python
labelz = ['sea', 'ground']
events_train, indices_train, label_train = get_labels_indices(path, labelz, patch_size)
```

```python
import numpy as np
np.isnan(events_train).sum()
```

```python
model, loss = fit_MLR(events_train, label_train, indices_train, tau_cla, patch_size)
```

```python
tau = tau_cla
sensor_size = (128,128,2)
surface_dimensions = sensor_size
radius_x, radius_y = 0, 0
events = events_train[20000:30000,:]
decay = "exp"

timestamp_memory = np.zeros(
    (sensor_size[2], sensor_size[1] + radius_y * 2, sensor_size[0] + radius_x * 2)
)
timestamp_memory -= tau * 3 + 1
all_surfaces = np.zeros(
    (len(events), sensor_size[2], surface_dimensions[1], surface_dimensions[0])
)
for index, event in enumerate(events):
    x = int(event[0])
    y = int(event[1])
    timestamp_memory[int(event[3]), y + radius_y, x + radius_x] = event[2]
    if radius_x > 0 and radius_y > 0:
        timestamp_context = (
            timestamp_memory[
                :, y : y + surface_dimensions[1], x : x + surface_dimensions[0]
            ]
            - event[2]
        )
    else:
        timestamp_context = timestamp_memory - event[2]

    if decay == "lin":
        timesurface = timestamp_context / (3 * tau) + 1
        timesurface[timesurface < 0] = 0
    elif decay == "exp":
        timesurface = np.exp(timestamp_context / tau)
    if timesurface.max()==np.inf:
        print(events[2],timestamp_context[0][0], timesurface[0][0], timestamp_memory[0][0])
    all_surfaces[index, :, :, :] = timesurface
#return all_surfaces
print(np.isnan(all_surfaces).sum())
```
